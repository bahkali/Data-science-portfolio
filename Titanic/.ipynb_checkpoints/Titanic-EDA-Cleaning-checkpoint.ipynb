# Titanic Survival Prediction

## 1. Problem Statement 

Base of the datasset from kaggel. we need to predict if a passenger survived the sinking of the Titanic or not.

## Data 

The data is divided into training set and test sets. 
- survival	
- pclass Ticket class	
- sex		
- Age		
- sibsp	# of siblings / spouses aboard the Titanic	
- parch	# of parents / children aboard the Titanic	
- ticket =>Ticket number	
- fare	Passenger fare	
- cabin Cabin number	
- embarked

## Metric 

this problem use a supervised Learninig approach so the metric would be accuracy, precision and recall


```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
df = pd.read_csv('datasets/train.csv')
df.head()
```

```python

drop_feat = ['Name', 'PassengerId']
df = df.drop(drop_feat, axis=1)
df.head()
```

```python
df.Sex.unique()
```

```python
#convert sex to numerical

gender_num = { 'male': 1, 'female': 0}

df['Sex'] = df['Sex'].map(gender_num)
df.head()
```

```python
df.Cabin.unique()
```

```python
df.isna().sum()
```

```python
df.groupby(df['Cabin'].isnull())['Survived'].mean()
```

```python
# store 1 if the cabin of a row is null and 0 otherwise
df['Cabin_ind'] = np.where(df['Cabin'].isnull(), 0, 1)
df.head()
```

```python
df.drop(['Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python

df.groupby(df['Age'].isnull())['Survived'].mean()
```

```python
# REplace Age missing value to avg. value
df['Age_clean'] = df['Age'].fillna(df['Age'].mean())
df.head()
```

```python
df.isna().sum()
```

```python
df.drop('Age', axis=1, inplace=True)
```

```python
df.head()
```

```python
# combine SibSp and Parch to Family counts
df['family_cnt'] = df['SibSp'] + df['Parch']
df.head()
```

```python
df.corr()
```

```python
df.drop(['SibSp', 'Parch'], axis=1, inplace=True)
df.head()
```

```python
# sava this clean data before scaling  and scale the data for difference 
df_clean = df.to_csv('datasets/train-clean.csv')
```
